Cleaning Data

```python
def wilayah_clean(column_series):
    """
    Membersihkan dan menstandarkan nama wilayah dalam sebuah Series Pandas.
    """
    cleaned_series = column_series.astype(str).str.lower()
    cleaned_series = cleaned_series.str.replace('_', ' ', regex=True)
    cleaned_series = cleaned_series.str.replace(r'(kota|kab)\s*\.?\s*adm\.?\s*', '', regex=True)
    cleaned_series = cleaned_series.apply(lambda x: re.sub(r'\s*\d+$', '', x))
    cleaned_series = cleaned_series.str.strip()
    cleaned_series = cleaned_series.str.replace(r'\s+', ' ', regex=True)
    cleaned_series = cleaned_series.str.title()

    return cleaned_series
```

```python
dataset_campak['wilayah'] = wilayah_clean(dataset_campak['wilayah'])
dataset_campak['kecamatan'] = wilayah_clean(dataset_campak['kecamatan'])
dataset_campak.sample(3)
```

```python
dataset_dbd['wilayah'] = wilayah_clean(dataset_dbd['wilayah'])
dataset_dbd['kecamatan'] = wilayah_clean(dataset_dbd['kecamatan'])
dataset_dbd.sample(3)
```

```python
dataset_pnemonia['wilayah'] = wilayah_clean(dataset_pnemonia['wilayah'])
dataset_pnemonia['kecamatan'] = wilayah_clean(dataset_pnemonia['kecamatan'])
dataset_pnemonia.sample(3)
```

```python
dataset_thypoid['wilayah'] = wilayah_clean(dataset_thypoid['wilayah'])
dataset_thypoid['kecamatan'] = wilayah_clean(dataset_thypoid['kecamatan'])
dataset_thypoid.sample(3)
```

```python
def replace_value_kecamatan(df, old_value, new_value):
    df['kecamatan'] = df['kecamatan'].str.replace(old_value, new_value)
    return df
```
